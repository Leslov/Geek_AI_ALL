```python
from matplotlib import pyplot as plt
import numpy as np
import math
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
t = np.linspace(-3, 3, num=51)
t
```

```python
y1 = [x**2 for x in t]
y2 = [2 * x + 0.5 for x in t]
y3 = [ -3 * x - 1.5 for x in t]
y4 = [math.sin(x) for x in t]
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))
fig.tight_layout(pad=0.3)
ys = [y1, y3, y2, y4]
for row in ax:
    for col in row:
        col.plot(t, ys.pop())
ax[0][1].set_xlim(-5, 5)
```
