```python
from matplotlib import pyplot as plt
import numpy as np
import math
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
t = np.linspace(0, 10, num=51)
t
```

```python
f = [math.cos(x) for x in t]
```

```python
fig = plt.figure()
plt.xlim(0.5, 9.5)
plt.ylim(-2.5, 2.5)
plt.xlabel('Значения t', fontsize=18)
plt.ylabel('Значения f', fontsize=18)
plt.plot(t, f, color='green')
fig.suptitle('foo0', fontsize=18)
plt.show()
```
