```python
from matplotlib import pyplot as plt
import numpy as np
import math
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
t = np.linspace(0, 10, num=51)
t
```

```python
f = [math.cos(x) for x in t]
```

```python
plt.plot(t, f, color='green')
fig = plt.figure

del plt.xlabel
plt.xlabel('foo', fontsize=18)
plt.show()
```
